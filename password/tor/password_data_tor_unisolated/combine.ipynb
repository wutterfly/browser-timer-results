```python
# imports and config
import pandas as pd
```

```python
# all parts of the complete dataset
dataframes = [
    pd.read_csv("./text-timing-data(0).csv"),
    pd.read_csv("./text-timing-data(1).csv"),
    pd.read_csv("./text-timing-data(2).csv"),
    pd.read_csv("./text-timing-data(3).csv"),
    pd.read_csv("./text-timing-data(4).csv"),
    pd.read_csv("./text-timing-data(5).csv"),
    pd.read_csv("./text-timing-data(6).csv"),
    pd.read_csv("./text-timing-data(7).csv"),
    pd.read_csv("./text-timing-data(8).csv"),
    pd.read_csv("./text-timing-data(9).csv"),
    pd.read_csv("./text-timing-data(10).csv"),
    pd.read_csv("./text-timing-data(11).csv"),
    pd.read_csv("./text-timing-data(12).csv"),
    pd.read_csv("./text-timing-data(13).csv"),
    pd.read_csv("./text-timing-data(14).csv"),
    pd.read_csv("./text-timing-data(15).csv"),
    pd.read_csv("./text-timing-data(16).csv"),
    pd.read_csv("./text-timing-data(17).csv"),
    pd.read_csv("./text-timing-data(18).csv"),
    pd.read_csv("./text-timing-data(19).csv"),
    pd.read_csv("./text-timing-data(20).csv"),
]
assert len(dataframes) == 21
```

```python
# combine parts into complete dataset
complete = pd.concat(dataframes)


# read data from simulation
simulation_data = pd.read_csv("./password_data_rs.csv")
```

```python
# specify number of events per password input
n = 2 * 11

# check that lengths are equal
original_len = 20400
is_len = len(complete) / n
assert is_len  == original_len == len(simulation_data)

# group all events for one password input into individual chunks
list_df = [complete[i:i+n]
           for i in range(0, len(complete), n)]
list_df
```

```python
def construct_row(row: pd.DataFrame, meta: pd.Series):
    """
    Construct one row (one password input) from individual events.
    """

    # idividual event timestamps | [0] -> keydown | [1] -> keyup
    period = [-1, -1]
    t = [-1, -1]
    i = [-1, -1]
    e = [-1, -1]
    five = [-1, -1]
    shift = [-1, -1]
    r = [-1, -1]
    shiftR = [-1, -1]
    o = [-1, -1]
    a = [-1, -1]
    n = [-1, -1]
    l = [-1, -1]
    ret = [-1, -1]

    # extract idividual event timestamps
    for index, row in row.iterrows():
        # .
        if row['key'] == '.' and row['type'] == 'keydown':
            assert period[0] == -1
            period[0] = row['timestamp']

        if row['key'] == '.' and row['type'] == 'keyup':
            assert period[1] == -1
            period[1] = row['timestamp']

        # t
        if row['key'] == 't' and row['type'] == 'keydown':
            assert t[0] == -1
            t[0] = row['timestamp']

        if row['key'] == 't' and row['type'] == 'keyup':
            assert t[1] == -1
            t[1] = row['timestamp']

        # i
        if row['key'] == 'i' and row['type'] == 'keydown':
            assert i[0] == -1
            i[0] = row['timestamp']

        if row['key'] == 'i' and row['type'] == 'keyup':
            assert i[1] == -1
            i[1] = row['timestamp']

        # e
        if row['key'] == 'e' and row['type'] == 'keydown':
            assert e[0] == -1
            e[0] = row['timestamp']

        if row['key'] == 'e' and row['type'] == 'keyup':
            assert e[1] == -1
            e[1] = row['timestamp']

        # 5
        if row['key'] == '5' and row['type'] == 'keydown':
            assert five[0] == -1
            five[0] = row['timestamp']

        if row['key'] == '5' and row['type'] == 'keyup':
            assert five[1] == -1
            five[1] = row['timestamp']

        # Shift
        if row['key'] == 'Shift' and row['type'] == 'keydown':
            assert shift[0] == -1
            shift[0] = row['timestamp']

        if row['key'] == 'Shift' and row['type'] == 'keyup':
            assert shift[1] == -1
            shift[1] = row['timestamp']

        # r
        if (row['key'] == 'r' or row['key'] == 'R') and row['type'] == 'keydown':
            assert r[0] == -1
            r[0] = row['timestamp']

        if (row['key'] == 'r' or row['key'] == 'R') and row['type'] == 'keyup':
            assert r[1] == -1
            r[1] = row['timestamp']

        # o
        if row['key'] == 'o' and row['type'] == 'keydown':
            assert o[0] == -1
            o[0] = row['timestamp']

        if row['key'] == 'o' and row['type'] == 'keyup':
            assert o[1] == -1
            o[1] = row['timestamp']

        # a
        if row['key'] == 'a' and row['type'] == 'keydown':
            assert a[0] == -1
            a[0] = row['timestamp']

        if row['key'] == 'a' and row['type'] == 'keyup':
            assert a[1] == -1
            a[1] = row['timestamp']

        # n
        if row['key'] == 'n' and row['type'] == 'keydown':
            assert n[0] == -1
            n[0] = row['timestamp']

        if row['key'] == 'n' and row['type'] == 'keyup':
            assert n[1] == -1
            n[1] = row['timestamp']

        # l
        if row['key'] == 'l' and row['type'] == 'keydown':
            assert l[0] == -1
            l[0] = row['timestamp']

        if row['key'] == 'l' and row['type'] == 'keyup':
            assert l[1] == -1
            l[1] = row['timestamp']

        # return
        if (row['key'] == 'Enter' or row['key'] == 'Return') and row['type'] == 'keydown':
            assert ret[0] == -1
            ret[0] = row['timestamp']

        if (row['key'] == 'Enter' or row['key'] == 'Return') and row['type'] == 'keyup':
            assert ret[1] == -1
            ret[1] = row['timestamp']

    # check that all events were captured
    assert period[0] != -1 and period[1] != -1
    assert t[0] != -1 and t[1] != -1
    assert i[0] != -1 and i[1] != -1
    assert e[0] != -1 and e[1] != -1
    assert five[0] != -1 and five[1] != -1
    # assert shift[0] != -1 and shift[1] != -1 // does not get recorded in tor browser!
    assert r[0] != -1 and r[1] != -1
    assert o[0] != -1 and o[1] != -1
    assert a[0] != -1 and a[1] != -1
    assert n[0] != -1 and n[1] != -1
    assert l[0] != -1 and l[1] != -1
    assert ret[0] != -1 and ret[1] != -1

    # ground timestamps on first keypress (. keydown)
    start = period[0]

    period[0] = (period[0] - start) / 1000
    assert period[0] == 0
    period[1] = (period[1] - start) / 1000
    t[0] = (t[0] - start) / 1000
    t[1] = (t[1] - start) / 1000
    i[0] = (i[0] - start) / 1000
    i[1] = (i[1] - start) / 1000
    e[0] = (e[0] - start) / 1000
    e[1] = (e[1] - start) / 1000
    five[0] = (five[0] - start) / 1000
    five[1] = (five[1] - start) / 1000
    shift[0] = (shift[0] - start) / 1000
    shift[1] = (shift[1] - start) / 1000
    r[0] = (r[0] - start) / 1000
    r[1] = (r[1] - start) / 1000
    shiftR[0] = r[0]
    shiftR[1] = r[1]
    o[0] = (o[0] - start) / 1000
    o[1] = (o[1] - start) / 1000
    a[0] = (a[0] - start) / 1000
    a[1] = (a[1] - start) / 1000
    n[0] = (n[0] - start) / 1000
    n[1] = (n[1] - start) / 1000
    l[0] = (l[0] - start) / 1000
    l[1] = (l[1] - start) / 1000
    ret[0] = (ret[0] - start) / 1000
    ret[1] = (ret[1] - start) / 1000

    subject = meta['subject']
    session = meta['session']
    rep = meta['rep']

    data = {
        "subject": subject,
        "sessionIndex": session,
        "rep": rep,
        "H.period": round(period[1] - period[0], 4),      # . down -> . up
        "DD.period.t": round(t[0] - period[0], 4),        # . down -> t down
        "UD.period.t": round(t[0] - period[1], 4),        # . up -> t down
        "H.t": round(t[1] - t[0], 4),                     # t down -> t up
        "DD.t.i": round(i[0] - t[0], 4),                  # t down -> i down
        "UD.t.i": round(i[0] - t[1], 4),                  # t up -> i down
        "H.i": round(i[1] - i[0], 4),                     # i down -> i up
        "DD.i.e": round(e[0] - i[0], 4),                  # i down -> e down
        "UD.i.e": round(e[0] - i[1], 4),                  # i up -> e down
        "H.e": round(e[1] - e[0], 4),                     # e down -> down
        "DD.e.five": round(five[0] - e[0], 4),            # e down -> 5 down
        "UD.e.five": round(five[0] - e[1], 4),            # e up -> 5 down
        "H.five": round(five[1] - five[0], 4),            # 5 down -> 5 up
        # 5 down ->  ShiftR down
        "DD.five.Shift.r": round(shiftR[0] - five[0], 4),
        # 5 up -> ShiftR down
        "UD.five.Shift.r": round(shiftR[0] - five[1], 4),
        # ShiftR down -> ShiftR up
        "H.Shift.r": round(shiftR[1] - shiftR[0], 4),
        # ShiftR down -> o down
        "DD.Shift.r.o": round(o[0] - shiftR[0], 4),
        "UD.Shift.r.o": round(o[0] - shiftR[1], 4),       # ShiftR up -> o down
        "H.o": round(o[1] - o[0], 4),                    # o down -> o up
        "DD.o.a": round(a[0] - o[0], 4),                  # o down -> a down
        "UD.o.a": round(a[0] - o[1], 4),                 # o up -> a down
        "H.a": round(a[1] - a[0],    4),                 # a down -> a up
        "DD.a.n": round(n[0] - a[0],  4),                # a down -> n down
        "UD.a.n": round(n[0] - a[1], 4),                 # a up -> n down
        "H.n": round(n[1] - n[0],   4),                  # n down -> n up
        "DD.n.l": round(l[0] - n[0],  4),                # n down -> l down
        "UD.n.l": round(l[0] - n[1],  4),                # n up -> l down
        "H.l": round(l[1] - l[0],   4),                  # l down -> l up
        # l down -> Return down
        "DD.l.Return": round(ret[0] - l[0], 4),
        "UD.l.Return": round(ret[0] - l[1], 4),           # l up -> Return down
        # Return down -> Return up
        "H.Return": round(ret[1] - ret[0],  4)
    }

    return data
```

```python
# reconstruct all rows
data_list = []
for i in range(0, len(list_df)):
    row = construct_row(list_df[i], simulation_data.iloc[i])
    data_list.append(row)

assert len(data_list) == original_len
```

```python
# save reconstructed dataset
full_data_set = pd.DataFrame(data_list)

full_data_set.to_csv("./full_data_set.csv", index=False)
```
