```python
# convert absolute timestamps to relative timestamps
# normal 50 
grounded_browser_natural_normal_50 = []
grounded_os_natural_normal_50 = []
grounded_browser_machine_normal_50 = []
grounded_os_machine_normal_50 = []

# normal 75 
grounded_browser_natural_normal_75 = []
grounded_os_natural_normal_75 = []
grounded_browser_machine_normal_75 = []
grounded_os_machine_normal_75 = []

# normal 100
grounded_browser_natural_normal_100 = []
grounded_os_natural_normal_100 = []
grounded_browser_machine_normal_100 = []
grounded_os_machine_normal_100 = []

# abovenormal 75
grounded_browser_natural_abovenormal_75 = []
grounded_os_natural_abovenormal_75 = []
grounded_browser_machine_abovenormal_75 = []
grounded_os_machine_abovenormal_75 = []

# ground event timestamps on first event
for i in range(0, 200):
    # normal 50
    grounded_browser_natural_normal_50.append(browser_data_natural_normal_50[i] - browser_data_natural_normal_50[0])
    grounded_os_natural_normal_50.append(os_data_natural_normal_50[i] - os_data_natural_normal_50[0])
    grounded_browser_machine_normal_50.append(browser_data_machine_normal_50[i] - browser_data_machine_normal_50[0])
    grounded_os_machine_normal_50.append(os_data_machine_normal_50[i] - os_data_machine_normal_50[0])

    # normal 75
    grounded_browser_natural_normal_75.append(browser_data_natural_normal_75[i] - browser_data_natural_normal_75[0])
    grounded_os_natural_normal_75.append(os_data_natural_normal_75[i] - os_data_natural_normal_75[0])
    grounded_browser_machine_normal_75.append(browser_data_machine_normal_75[i] - browser_data_machine_normal_75[0])
    grounded_os_machine_normal_75.append(os_data_machine_normal_75[i] - os_data_machine_normal_75[0])

    # normal 100
    grounded_browser_natural_normal_100.append(browser_data_natural_normal_100[i] - browser_data_natural_normal_100[0])
    grounded_os_natural_normal_100.append(os_data_natural_normal_100[i] - os_data_natural_normal_100[0])
    grounded_browser_machine_normal_100.append(browser_data_machine_normal_100[i] - browser_data_machine_normal_100[0])
    grounded_os_machine_normal_100.append(os_data_machine_normal_100[i] - os_data_machine_normal_100[0])

    # abovenormal 75
    grounded_browser_natural_abovenormal_75.append(browser_data_natural_abovenormal_75[i] - browser_data_natural_abovenormal_75[0])
    grounded_os_natural_abovenormal_75.append(os_data_natural_abovenormal_75[i] - os_data_natural_abovenormal_75[0])
    grounded_browser_machine_abovenormal_75.append(browser_data_machine_abovenormal_75[i] - browser_data_machine_abovenormal_75[0])
    grounded_os_machine_abovenormal_75.append(os_data_machine_abovenormal_75[i] - os_data_machine_abovenormal_75[0])
```

```python
complete = pd.DataFrame({
    'browser_natural_normal_50': grounded_browser_natural_normal_50,
    'os_natural_normal_50': grounded_os_natural_normal_50,
    'browser_machine_normal_50': grounded_browser_machine_normal_50,
    'os_machine_normal_50': grounded_os_machine_normal_50,

    'browser_natural_normal_75': grounded_browser_natural_normal_75,
    'os_natural_normal_75': grounded_os_natural_normal_75,
    'browser_machine_normal_75': grounded_browser_machine_normal_75,
    'os_machine_normal_75': grounded_os_machine_normal_75,

    'browser_natural_normal_100': grounded_browser_natural_normal_100,
    'os_natural_normal_100': grounded_os_natural_normal_100,
    'browser_machine_normal_100': grounded_browser_machine_normal_100,
    'os_machine_normal_100': grounded_os_machine_normal_100,

    'browser_natural_abovenormal_75': grounded_browser_natural_abovenormal_75,
    'os_natural_abovenormal_75': grounded_os_natural_abovenormal_75,
    'browser_machine_abovenormal_75': grounded_browser_machine_abovenormal_75,
    'os_machine_abovenormal_75': grounded_os_machine_abovenormal_75,

    'type': types,
})

complete.to_csv("./complete.csv", index=False)
```
