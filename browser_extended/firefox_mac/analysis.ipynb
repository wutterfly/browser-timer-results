```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# read datset
data = pd.read_csv('./complete.csv')

# get natural data
natural_os = data['os_natural'].to_list()
natural_browser = data['browser_natural'].to_list()
```

```python
# calculate distances

def distances(timestamps:list) -> list[float]:
    last = None
    distances = []
    for x in timestamps:
        if last is None:
            last = x
            distances.append(0)
            continue
    
        else:
            assert (x - last) > 0
            distances.append(x - last)
            last = x

    return distances
```

```python
natural_os_distances = distances(natural_os)
natural_browser_distances = distances(natural_browser)
```

```python
plt.plot(natural_os_distances, color="blue", label="OS Distances | Keyboard Input", linewidth=1.0)
plt.plot(natural_browser_distances, color="red",label="Browser Distances | Keyboard Input", linewidth=1.0)

#plt.plot(machine_os_distances, color="green", label="OS Distances | Script Input", linewidth=1.0)
#plt.plot(machine_browser_distances, color="yellow",label="Browser Distances | Script Input", linewidth=1.0)

plt.title('Distances to last event', fontsize=18)
plt.xlabel('Event', fontsize=18)
plt.ylabel('Distance (sec)', fontsize=18)
plt.legend()

plt.tight_layout()
plt.show()
```
