```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# calculate differences between browser and os data
def get_differences(browser: pd.Series, os: pd.Series) -> list:
    assert len(browser) == len(os)
    diffs = [] 
    for i in range(0, len(browser)):
        diffs.append(abs(browser.loc[i] - os.loc[i])  * 1000)

    return diffs
```

```python
def calc(diffs_natural, diffs_machine):
    diffs_natural = np.array(diffs_natural)
    diffs_machine = np.array(diffs_machine)

    mean = diffs_natural.mean()
    std = diffs_natural.std()
    min = diffs_natural.min()
    max = diffs_natural.max()

    r = 1

    print("Natural")
    print("Min : ", round(min, r))
    print("Max : ", round(max, r))
    print("Mean: ", round(mean, r))
    print("Std : ", round(std, r))

    mean = diffs_machine.mean()
    std = diffs_machine.std()
    min = diffs_machine.min()
    max = diffs_machine.max()

    print("Machine")
    print("Min : ", round(min, r))
    print("Max : ", round(max, r))
    print("Mean: ", round(mean, r))
    print("Std : ", round(std, r))
```

```python
# firefox stresstest | idle
firefox = pd.read_csv("./firefox_stresstest/complete.csv")
firefox_diffs_natural_idle = get_differences(firefox['browser_natural_idle'], firefox['os_natural_idle'])
firefox_diffs_machine_idle = get_differences(firefox['browser_machine_idle'], firefox['os_machine_idle'])
calc(firefox_diffs_natural_idle,firefox_diffs_machine_idle)

#plt.title('Timing differences between input events in Browser and native application (Firefox)', fontsize=18)
#plt.ylabel('Differences in seconds', fontsize=18)
#plt.xlabel('Event', fontsize=18)
#plt.plot(firefox_diffs_natural_idle, label= "Keyboard Input")
#plt.plot(firefox_diffs_machine_idle, label= "Script Input")
#plt.legend()
#plt.show()

# plot firefox
#plt.boxplot([firefox_diffs_natural_idle, firefox_diffs_machine_idle], labels=["Differences with manual input",
#            "Differences with machine input"], showfliers=False, showmeans=True)
#plt.title("Timing differences between input events in Browser and native application (Firefox | Stresstest idle)", fontsize=18)
#plt.xticks(fontsize=18)
#plt.xlabel("", fontsize=18)
#plt.ylabel("Differences in seconds", fontsize=18)
#plt.tight_layout()
#plt.savefig("./firefox_stresstest/fig_idle.jpeg", bbox_inches="tight", pad_inches=0.3)
#plt.show()
```
